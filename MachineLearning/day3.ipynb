# Machine Learning

### 과대 적합(overfitting)

train 데이터에 너무 과도하게 학습되어 train 데이터에만 잘 동작하고 test에서는 예측 성능이 저하되는 현상
너무 상세하고 복잡한 모델리을 하여 훈련데이터에만 과도하게 정확히 동작하는 모델
### 과소 적합(underfitting)

train 데이터가 충분히 반영되지 않은 형태
모델링을 너무 간단하게 하여 성능이 제대로 나오지 않은 모델



### 모델의 복잡도 곡선




### 모델의 복잡도 해결

 - 일반적으로 데이터 양이 많으면 일반화에 도움이 됨
 - 주어진 훈련데이터의 다양성 보장되어야 함
 - 다양한 데이터 포인트를 골고루 나타내기
 - 편중된 데이터를 많이 모으는 것은 도움이 되지 않음
 - 규제(Regularization)을 통해 모델의 복잡도를 적정선으로 결정



### K-Nearest Neighbors (knn)




장단점 및 키워드

- n_neighbors :이웃수
- 이해하기 며운 쉬운 모델, 큰 조정 없이도 나쁘지 않은 성능을 발휘하는 기초모델
- 새로운 테스트 데이터 세트가 들어오면 훈련데이터 세트와의 거리를 계산해서 훈련 데이터 세트가 크면 예측이 느려짐
- 거리를 측정하기 떄문에 데이터의 스케일 조정이 필요할  수 있음
- 직접적인 예측에 사용되기보다는 주로 데이터를 파악하기 위한 용도로 가볍게 사용
