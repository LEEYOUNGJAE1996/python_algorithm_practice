```python
import cv2 
print(cv2.__version__)
```

```python
# 차트 라이브러리를 통해서 이미지를 출력한다.
import matplotlib.pyplot as plt
# 이미지 로드하기
# imread() : 이미지를 읽어오는 함수
# cv2.IMREAD_COLOR
chun = cv2.imread('./img/chunsik.jpg',cv2.IMREAD_COLOR)

# 이미지 출력
plt.imshow(chun)
```

춘식이 색이 다른 이유



```python
#  cvtColor() : 이미지의 색상을 변환하는 함수
cvt_chun = cv2.cvtColor(chun, cv2.COLOR_BGR2RGB)
# 축의 눈금 삭제하기
plt.xticks([])
plt.yticks([])
# 이미지 출력
plt.imshow(cvt_chun)
```

```python
lena = cv2.imread('D:/workstation/python_algorithm_practice/OPENCV/img/lena2.jpg',cv2.IMREAD_COLOR)
cvt_lena = cv2.cvtColor(lena,cv2.COLOR_BGR2RGB)
plt.imshow(cvt_lena)
```

```python
### 각각의 BGR 정보를 나눠준 후에 다시 합쳐주는 방식으로 자리를 교체하는 것이 가능
lena = cv2.imread('D:/workstation/python_algorithm_practice/OPENCV/img/lena2.jpg',cv2.IMREAD_COLOR)
B,G,R = cv2.split(lena)
RGB_chan=cv2.merge([R,G,B])
plt.imshow(RGB_chan)
```

```python
# 새로운 창을 띄워서 이미지 출력하는 방법 - gray 스케일로 가져오기

# cv2를 사용해서 이미지 불러오기
lena = cv2.imread('./img/lena2.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('lena',lena)
# 창을 띄운 이후에 창을 유지시키는 시간 설정(키보드 입력 시에 창이 닫힐 수 있도록 설정
# cv2.waitKey(시간) 시간이 '0'인경우 키보드가 눌릴 때까지 무한정 대기
cv2.waitKey(3000)
# 창 닫기
cv2.destroyAllWindows()
# plt.imshow(lena)
```

### 코드 정리
- cv2.imread() : 이미지를 읽어오기
- cv2.cvtColor() : 이미지의 색상 공간을 변환
- cv2.imshow() : 새창에 이미지를 출력
- cv2.waitKey() : 키 입력을 대기하는 코드
