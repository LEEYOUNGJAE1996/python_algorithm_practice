###  (binary)  

-    0  1 .  0   1 
-       . , 
- cv2.threshold(img,,value,flag)
    - img :  
    - threshold :   (0 1   / )
    - value :     
    - flag :    (   )    
        - cv2.THRESH_BINARY :    value,  0 
        - cv2.THRESH_BINARY_INV :    0,  value  (inverse)
        - cv2.THRESH_TRUNC :    ,    
        - cv2.THRESH_TOZERO :    ,  0 
        - cv2.THRESH_TOZERO_INV :    0,   

```python
#   

gray_image = cv2.imread('./img/lena2.jpg', cv2.IMREAD_GRAYSCALE)
#   
#  2 
mat, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

#   
cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#  flag 
# subplot(m,n,p)      
# m * n   
# p :  ?
plt.subplot(2,2,1)

_,th1 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
plt.xticks([])
plt.yticks([])
plt.title('binary')
plt.imshow(th1,cmap ='gray' )
plt.subplot(2,2,2)
_,th2 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY_INV)
plt.xticks([])
plt.yticks([])
plt.title('binary_inv')
plt.imshow(th2,cmap ='gray' )
plt.subplot(2,2,3)
_,th3 = cv2.threshold(gray_img,127,255,cv2.THRESH_TRUNC)
plt.xticks([])
plt.yticks([])
plt.title('trunc')
plt.imshow(th3,cmap ='gray' )
plt.subplot(2,2,4)
_,th4 = cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO)
plt.xticks([])
plt.yticks([])
plt.title('TOZERO')
plt.imshow(th4,cmap ='gray' )
```

```python
color = cv2.imread('./img/lena2.jpg',cv2.IMREAD_COLOR)
color = cv2.cvtColor(color, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(color, cv2.COLOR_RGB2GRAY)
R,G,B = cv2.split(color)

_,R_th= cv2.threshold(R,127,255,cv2.THRESH_BINARY)
_,G_th= cv2.threshold(G,127,255,cv2.THRESH_BINARY)
_,B_th= cv2.threshold(B,127,255,cv2.THRESH_BINARY)

change_color = cv2.merge([R_th,G_th,B_th])
```
