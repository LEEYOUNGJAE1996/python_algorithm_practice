### 이진 (binary) 이미지 만들기

- 이진 이미지는 픽셀값이 0 또는 1로만 구성되어있다.  0 이 흑 1이 백
- 이진이미지는 배경과 객체의 구분이 필요할 때 사용한다. 관심영역, 비관심영역
- cv2.threshold(img,기준값,value,flag)
    - img : 이미지 변수
    - threshold : 픽셀 문턱값 (0과 1로 나누기 위한 기준/ 임계값)
    - value : 임계값보다 클 겨웅 적용할 값
    - flag : 문턱값 적용 방법 (다양한 형태의 이진화 방법) 실제로는 위의 두값을 사용
        - cv2.THRESH_BINARY : 픽셀값이 문턱값보다 크면 value, 아니면 0을 할당
        - cv2.THRESH_BINARY_INV : 픽셀값이 문턱값보다 크면 0, 아니면 value를 할당 (inverse)
        - cv2.THRESH_TRUNC : 픽셀값이 문턱값보다 크면 문턱값, 아니면 픽셀값을 그대로 할당
        - cv2.THRESH_TOZERO : 픽셀값이 문턱값보다 크면 픽셀값, 아니면 0을 할당
        - cv2.THRESH_TOZERO_INV : 픽셀값이 문턱값보다 크면 0, 아니면 픽셀값을 할당

```python
# 이미지 흑백으로 불러오기

gray_image = cv2.imread('./img/lena2.jpg', cv2.IMREAD_GRAYSCALE)
# 이진 이미지로 변환
# 결과값 2개로 반환
mat, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# 이진 이미지 출력
cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 여러개의 flag 설정
# subplot(m,n,p) 여러개의 이미지를 동시 출력  
# m * n 을 통한 행렬
# p : 어느 위치에?
plt.subplot(2,2,1)

_,th1 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
plt.xticks([])
plt.yticks([])
plt.title('binary')
plt.imshow(th1,cmap ='gray' )
plt.subplot(2,2,2)
_,th2 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY_INV)
plt.xticks([])
plt.yticks([])
plt.title('binary_inv')
plt.imshow(th2,cmap ='gray' )
plt.subplot(2,2,3)
_,th3 = cv2.threshold(gray_img,127,255,cv2.THRESH_TRUNC)
plt.xticks([])
plt.yticks([])
plt.title('trunc')
plt.imshow(th3,cmap ='gray' )
plt.subplot(2,2,4)
_,th4 = cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO)
plt.xticks([])
plt.yticks([])
plt.title('TOZERO')
plt.imshow(th4,cmap ='gray' )
```

## 임계값(문턱값) 결정 알고리즘 : OTSU 알고리즘
임계값을 사람이 변경하고 확인하는 방법으로는 한계가 존재


- 이진 이미지를 만들 때, 가장 중요한 요소 : 임계값 설정
- otsu 이진화 알고리즘을 사용하여 최적의 임계값을 찾아보기
    - 임의의 임계값을 지정 -> pixel 변경
    - 두 분류의 명암 분포를 구분
    - 분산이 가장 적은 값으로 임계값을 설정 ( 두 분류의 명암붙포가 가장 균일할 때의 임계값을 설정)


### OTSU 이진화 알고리즘 적용 방법

- 임계값을 '-1'로 설정 후 옵션 THRESHOLD_OTSU 라고 설정 후 이진화 진행
- 단점 : 모든 경우의 수를 조사 -> 연산량 증가

### 임계값 결정하기 : 적응형
- 이미지를 여러 영역으로 나눔
- 해당되는 영역에 따라서 각자 임계치를 설정(각각의 임계값을 결정)
- 각 영역별로 이진화를 진행 후 나중에 병합(영역의 개수는 사용자가 결정)





적응형 

- cv2.adaptiveThreshold(img, value, method, type_flag, block_size, C)
    - img : 입력할 이미지 데이터
    - value : 임계값을 만족하는 픽셀에 적용할 값
    - method : 임계값을 결정하는 방법 --> 2가지가 존재
        - 평균
        - Gaussian 분포
        - 차이점 - 평균(이웃의 픽셀 평균값으로 임계값을 결정 - 선명하지만 잡음이 존재) , 가우시안 분포(흐리게 보이지만 잡음을 줄인다.)
    - type_flag : 적용 방법 
    - block_size : 영역을 나눌 블록의 개수 - 사용자 설정(m*n)  홀수개로 설정
    - C : 계산된 임계값 결과의 가감할 값 -> 균일화 목적

```python
# 스도쿠 영역별로 다르다.

gray_img = cv2.imread('./img/sample.jpg',cv2.IMREAD_GRAYSCALE)
threshold1 , binary = cv2.threshold(gray_img,127,255, cv2.THRESH_BINARY)
threshold2 , binary_ots = cv2.threshold(gray_img,-1,255, cv2.THRESH_BINARY|cv2.THRESH_OTSU)
# 적응형 평균, 가우시안
# cv2.adaptiveThreshold(img, value, method, type_flag, block_size, C)
# 평균
binary_adaptive_mean = cv2.adaptiveThreshold(gray_img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,5)
binary_adaptive_gaussian = cv2.adaptiveThreshold(gray_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,9,5)
plt.subplot(3,2,1)
plt.title('normal')
plt.imshow(gray_img,cmap='gray')
plt.subplot(3,2,2)
plt.title('original(127)')
plt.imshow(binary,cmap='gray')
plt.subplot(3,2,3)
plt.title(f'OTSU : {threshold2}')
plt.imshow(binary_ots,cmap='gray')
plt.subplot(3,2,4)
plt.title(f'OTSU : {threshold2}')
plt.imshow(binary_ots,cmap='gray')
plt.subplot(3,2,5)
# 평균 선명하지만 잡음이 많은편
plt.title('adaptive_mean')
plt.imshow(binary_adaptive_mean,cmap='gray')
plt.subplot(3,2,6)
# 선명하진 않지만 잡음 감소
plt.title('adaptive_gaussian')
plt.imshow(binary_adaptive_gaussian,cmap='gray')
```
