#  ROI에 도형 그리기

- cv2.line(이미지, 시작점 좌표, 끝 점 좌표, 색상, 선두께) --> 직선 그리기
- cv2.rectangle(이미지, 좌상단 좌표 , 우하단 좌표, 색상, 선두께) -> 사각형
- cv2.circle(이미지 , 중심점, 거리, 색상 , 선 두께)

```python
line_img = cv2.line(cvt_img,pt1= (300,200),pt2= (300,400), color=(0,255,0), thickness=10)
plt.imshow(line_img)
```

```python
circle_img = cv2.circle(cvt_img,center= (300,200),radius=50, color=(0,0,255), thickness=10)
plt.imshow(line_img)
```

```python
rectangle_img = cv2.rectangle(cvt_img,pt1= (200,300),pt2=(400,400), color=(255,0,0), thickness=10)
plt.imshow(rectangle_img)
```

```python
rectangle_img = cv2.(cvt_img,pt1= (200,300),pt2=(400,400), color=(255,0,0), thickness=10)
plt.imshow(rectangle_img)
```
