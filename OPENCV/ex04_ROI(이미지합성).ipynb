#  ROI에 도형 그리기

- cv2.line(이미지, 시작점 좌표, 끝 점 좌표, 색상, 선두께) --> 직선 그리기
- cv2.rectangle(이미지, 좌상단 좌표 , 우하단 좌표, 색상, 선두께) -> 사각형
- cv2.circle(이미지 , 중심점, 거리, 색상 , 선 두께)

```python
line_img = cv2.line(cvt_img,pt1= (300,200),pt2= (300,400), color=(0,255,0), thickness=10)
plt.imshow(line_img)
```

```python
circle_img = cv2.circle(cvt_img,center= (300,200),radius=50, color=(0,0,255), thickness=-1)
plt.imshow(line_img)
```

```python
rectangle_img = cv2.rectangle(cvt_img,pt1= (200,300),pt2=(400,400), color=(255,0,0), thickness=-1)
plt.imshow(rectangle_img)
```

### 이미지에 텍스트 작성하기! 

- cv2. putText(이미지, 출력내용_문자열, 출력시 시작 좌표, 폰트, 크기 , 색상, 굵기) 

```python
image = cv2.imread('./img/lena2.jpg')
cvt_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
text_img = cv2.putText(cvt_img,'Hi! Lena.', (120,100),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3)
plt.imshow(text_img)
```

### 한글 작성하기
- 1. 폰트 가져오기 
- 2. 이미지(넘파이 배열 형태) -> 자체 이미지 형태로 바꾼다. _ python library 사용 (PIL)
- 3. 이미지를 그리기
- 4. 그린 이미지에 폰트를 올려주기
- 5. 텍스트가 포함된 이미지를 다시 넘파이 배열화 진행!


font = ImageFont.truetype("fonts/gulim.ttc", 30) : 사용할 폰트와 크기를 가져온다

img_pil = Image.fromarray(img) : array 이미지를 PIL 이미지로 변환

draw = ImageDraw.Draw(img_pil) : PIL 이미지를 그림

draw.text((200, 20), "레나 이미지", font=font, fill=(0, 0, 255, 0)) : 이미지에 텍스트를 그림

img = np.array(img_pil) : PIL 이미지를 numpy 배열로 변환


```python
# 이미지화를 도와주는 라이브러리 불러오기
from PIL import ImageFont, ImageDraw, Image
import numpy as np
image = cv2.imread('./img/lena2.jpg')
cvt_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
font = ImageFont.truetype('./font/GULIM.TTC',50)
# 넘파이 배열 형태의 이미지를 이미지 형태로 변경
img_pil = Image.fromarray(cvt_img)
# 이미지 그려주기 
draw_img = ImageDraw.Draw(img_pil)
# 글씨쓰기
draw_img.text((40, 20), "레나 이미지", font=font, fill=(0, 0, 255, 0)) 
# 넘파이 배열화 진행
img = np.array(img_pil)
plt.imshow(img)
```
