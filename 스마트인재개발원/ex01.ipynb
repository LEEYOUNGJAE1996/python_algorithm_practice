### 인덱싱(indexing)
- 하나를 지정
### 슬라이싱(slicing)
- 범위를 지정하여 여러개를 범위로 가져온다.



```python
name = 'my name is young'
# 인덱싱 변수명[0]
# 주의사항 띄어쓰기도 인덱스를 가짐
print(name[-4] , name[-1])

# 슬라이싱 [start : end +1]
print(name[:])
print(name[11:])
print(name[:2])
print(name[::2])
```

```python
date_info = "20230209Sunny"
print('year : ' , date_info[0:4] )
print('month & day : ' , date_info[4:8] )
print('weather : ' , date_info[8:] )
# 문자열 format
# %기호 포매팅 - 데이터 타입에 따른 기호를 알아야한다.
month = 3 
day = 14
today = '오늘은 %d월 %d일 입니다.'%(month,day)
print(today)

# format 메서드 사용  
todayFormat = '오늘은 {}월 {}일 입니다.'.format(month,day)
print('format : ',todayFormat)
# f문자열 포맷팅
print(f'year : {date_info[0:4]} \nmonth & day : {date_info[4:8]}\nweather : {date_info[8:]}' )
```

```python
x , y  = 100 ,200
print("%d와 %d의 합은 %d입니다."%(x,y,x+y))
print(f"{x}와 {y}의 합은 {x+y}입니다.")
print("{0}와 {1}의 합은 {2}입니다.".format(x,y,x+y))
```
