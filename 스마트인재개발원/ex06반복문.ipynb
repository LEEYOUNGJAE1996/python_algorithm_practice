# 반복문
-프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문

for - 반복 횟수가 명확할 떄
while - 반복 횟수가 명확하지 않을 


# while


```python
num = 0
str1='파이썬 강의' 
while num < 13:
    print(str(num+1) +str1)
    num +=1
```

```python
while True:
    num1 = int(input('더할 첫번째 수  > '))
    num2 = int(input('더할 두번째 수  > '))
    if num1 == 0 and num2 == 0:
        break
    else :
        print(num1 + num2)
```

```python
import random as rd

num = rd.randrange(1,100)
game_on = True
count = 0
while game_on:
    user = int(input('enter the number'))
    if user > num : 
        print('lower')
        count += 1
    elif user < num:
        print('higher')
        count += 1
    else:
        print('correct')
        print(f'computer random number : {num}')
        print(f'user number : {user}')
        print(f'The number of times you tried : {count}')
        game_on = False
```

# for 구문
- 명확하게 반복의 횟수가 정해져 있을 경우 사용
- for 변수 in 문자열, 리스트 ---> 배열 형태를 가지면 인덱스로 값을 불러 올 수 있는 데이터 구조 

```python
foods = ['햄버거','라면','한솥','국밥']
for food in foods:
    print(food)
print(rd.choice(foods))
```

```python
list_ex = [1,2,3,4,5,6]
for num in list_ex:
    print(num,end=' ')
```

```python
score_list = [90,45,70,60,55]

count = 1
for score in score_list:
    if score >= 60:
        print(f'{count}번 학생은 합격입니다')
    else:
        print(f'{count}번 학생은 불합격입니다')
    count +=1
        
```

# for과 함꼐 사용하는 range() 메서드
-필요한 숫자리스트를 만들어내느 기능
- range(시작할 숫자(default = 0), 끝 +1, 증감(default=1))

```python
for i in range(10):
    print(i+1 ,end=' ')
print()
for i in range(4,10):
    print(i+1 ,end=' ')
print()
for i in range(1,10,2):
    print(i+1 ,end=' ')
```

```python
for i in range(97,76,-1):
    print(i,end=' ')
print()
for i in range(23,40,1):
    print(i,end=' ')
```
