# while


```python
num = 0
str1=' ' 
while num < 13:
    print(str(num+1) +str1)
    num +=1
```

```python
while True:
    num1 = int(input('    > '))
    num2 = int(input('    > '))
    if num1 == 0 and num2 == 0:
        break
    else :
        print(num1 + num2)
```

```python
import random as rd

num = rd.randrange(1,100)
game_on = True
count = 0
while game_on:
    user = int(input('enter the number'))
    if user > num : 
        print('lower')
        count += 1
    elif user < num:
        print('higher')
        count += 1
    else:
        print('correct')
        print(f'computer random number : {num}')
        print(f'user number : {user}')
        print(f'The number of times you tried : {count}')
        game_on = False
```

```python
foods = ['','','','']
for food in foods:
    print(food)
print(rd.choice(foods))
```

```python
list_ex = [1,2,3,4,5,6]
for num in list_ex:
    print(num,end=' ')
```

```python
score_list = [90,45,70,60,55]

count = 1
for score in score_list:
    if score >= 60:
        print(f'{count}  ')
    else:
        print(f'{count}  ')
    count +=1
        
```

```python
for i in range(10):
    print(i+1 ,end=' ')
print()
for i in range(4,10):
    print(i+1 ,end=' ')
print()
for i in range(1,10,2):
    print(i+1 ,end=' ')
```
