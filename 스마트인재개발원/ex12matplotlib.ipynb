```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# line chart 그려보기
y = [2,4,6,8]
plt.plot(y)
plt.show()
```

```python
# 데이터 x,y
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y,ls = ':',lw = 5, color ='m',marker = 'o',mfc = 'k')

# 차트 속성값
# ls (lineStyle)
# lw (lineWidth)
# color = b(blue),g(green),r(red),c(cyan),m(magenta),y(yellow),k(black),w(white) - line,marker 같이 변경
# 데이터의 위치를 표시해줄 수 있는 marker 설정 가능 - marker 
# marker의 내부색 설정 - mfc 
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
y = np.random.randint(1,10,8)
plt.plot(x,y,ls = '--', marker ='o',color = 'g')
plt.show()
```

```python
plt.plot(x,y,ls = ':', marker ='D',mfc='b',color = 'r')
plt.show()
```

```python
# marker 스타일 사용자 지정 '$원하는 모형$'
plt.plot(x,y,ls = ':', marker ='$D$',mfc='b',color = '#c79fef')
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', c = 'y', lw = 5, marker = '*', ms = 15, mec = 'c', mfc = 'w', mew = 2)
plt.show()
```

```python
# 격자를 넣을 필요가 존재
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', c = 'y', lw = 5, marker = '*', ms = 15, mec = 'c', mfc = 'w', mew = 2)
# grid 그리는 법 ( 그리드 : 격자눈금)
plt.grid(ls = ':') # 모든 차트설정은 plt.show()이전에 작성해야함
plt.show()
```
