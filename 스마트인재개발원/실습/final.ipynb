```python
import pandas as pd
```

```python
# 3개의 파일을 모두 읽어오기 
# 인덱스를 관서명
data2019 = pd.read_csv('2019.csv',encoding='euc-kr', index_col='관서명')
data2020 = pd.read_csv('2020.csv',encoding='euc-kr', index_col='관서명')
data2021 = pd.read_csv('2021.csv',encoding='euc-kr', index_col='관서명')
print(data2019)
print(data2020)
print(data2021)
```

```python
# 특정 연도에만 존재하는 결찰청을 삭제 (연산 시 NaN)
data2021.drop('광주지방경찰청',inplace=True)
```

```python
# 년도별 총 합을 구해야한다. (제일 우측에 새로운 컬럼)
data2019['총합'] = data2019.loc[:,'살인':'폭력'].sum(axis=1)
data2020['총합'] = data2020.loc[:,'살인':'폭력'].sum(axis=1)
data2021['총합'] = data2021.loc[:,'살인':'폭력'].sum(axis=1)
data2019
```

```python
# 구분이 '발생건수'인 값들만 가져오기 (boolean 인덱싱)
data2019b=data2019[data2019['구분'] == '발생건수']
data2020b=data2020[data2020['구분'] == '발생건수']
data2021b=data2021[data2021['구분'] == '발생건수']
print(data2019b)
print(data2020b)
print(data2021b)
```

```python
# # 총합을 기준으로 데이터 자르기 Series 형태로 출력
# series2019 = pd.Series(data2019b['총합'])
# series2020 = pd.Series(data2020b['총합'])
# series2021 = pd.Series(data2021b['총합'])
series2019 = data2019b['총합']
series2020 = data2020b['총합']
series2021 = data2021b['총합']
# data2020b['총합']
# data2021b['총합']
```

```python
# 자른 데이터 이름 변경해주기
series2019.name = '2019총계'
series2020.name = '2020총계'
series2021.name = '2021총계'
series2019
```

```python
# 증감율 계산 
# (금년 - 작년) / 작년 * 100
upDown2020 = (series2020 - series2019)/series2019 * 100
upDown2021 = (series2021 - series2020)/series2020 * 100
upDown2020.name = '2019-2020증감율'
upDown2021.name = '2020-2021증감율'
print(upDown2020)
print(upDown2021)
```

```python
# 데이터 병합 dataframe 화
result =pd.concat([series2019,upDown2020,series2020,upDown2021,series2021],axis=1)
```

```python
# 파일 내보내기
result.to_csv('2019-2021 범죄 증감률.csv')
```
